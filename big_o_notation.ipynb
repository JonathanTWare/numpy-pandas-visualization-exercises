```python
import matplotlib.pyplot as plt
import math as math
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize = (16,10))

x= (range(1,10))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4= [n * math.log(n) for n in x]
y5= [n**2 for n in x]
y6= [2**n for n in x]
y7= [math.factorial(n) for n in x]
y8= [n**n for n in x]


plt.plot(x,y1, color ='red', label='$y1= O(1)$')
plt.plot(x,y2, color ='blue', label='$y2 = O(log n)$')
plt.plot(x,y3, color ='purple', label='$y3= O(n)$')
plt.plot(x,y4, color='green', label='$y4 = O(n log n)$')
plt.plot(x,y5, color='cyan', label='$y5 = O(n^2)$')
plt.plot(x,y6, color='darkgreen', label='$y6 = O(2^n)$')
plt.plot(x,y7, color='orange', label='$y7 = O(n!)$')
plt.plot(x,y8, color='yellow', label='$y8 = O(n^n)$')

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation')
plt.xlim(1, 9)
plt.ylim(0, 70)


plt.legend()

plt.show()
```
